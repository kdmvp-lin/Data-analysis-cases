### 任务2.1

```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series
import matplotlib.pyplot as plt
```

```python
saleMapNum=pd.read_excel('./../result/result1_2.xlsx')
saleMapNum
```

```python
saleMapNum.fillna(method='ffill',axis=0,inplace=True)
```

```python
saleMapNum.sort_values(by='销售额',axis=0,ascending=False)
```

```python
#饼图
plt.title('sales volume',fontsize=50,verticalalignment='top')
plt.pie(saleMapNum['销售额'],radius=10,labels=saleMapNum['国家'],shadow=True,textprops={'fontsize':20,'color':'black'})
```

```python
plt.figure(figsize=(40,40))
plt.title('profit',fontsize=50,verticalalignment='top')
plt.tick_params(axis='both', labelsize=5)
plt.bar(saleMapNum['国家'],saleMapNum['利润'])
```
